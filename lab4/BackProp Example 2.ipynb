```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X1 = 1.5
X2 = -4
y = 0.5
```

```python
W1 = np.random.random()
W2 = np.random.random()
print("W1 = ",W1)
print("W2 = ",W2)
```

```python
lr = 0.05
epochs = 30

errorList = []
yList = []

for epoch in range(epochs):
    #Forward Pass
    y_pred = X1*W1 + X2*W2
    yList.append(y_pred)

    #Compute Error
    error = np.square(y-y_pred)
    errorList.append(y_pred)

    #Compute Gradients
    gradient1 = -2*X1*(y-y_pred)
    gradient2 = -2*X2*(y-y_pred)

    #Update the weights
    W1 -= lr*gradient1
    W2 -= lr*gradient2
```

```python
plt.plot(range(epochs), errorList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
plt.savefig("Epoch vs Error.png")
plt.xticks(range(epochs));
```

```python
plt.plot(range(epochs), yList)
plt.xlabel("Epochs")
plt.ylabel("y_pred")
plt.title("Epochs vs y_pred")
plt.savefig("Epoch vs y_pred.png")
```
