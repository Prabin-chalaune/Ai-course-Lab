```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = 2.5
y = 0.5
```

```python
W = -0.5
```

```python
lr = 0.1
epochs = 20
yList = []
errorList = []
wList = []

for epoch in range(epochs):
    #Forward Pass
    y_pred = X*W

    #Compute the error
    error = np.square(y-y_pred)

    #Compute the gradieny
    gradient = -2*(y - y_pred) * X

    #Update the weight
    W = W - lr*gradient

    wList.append(W)
    yList.append(y_pred)
    errorList.append(error)
```

```python
plt.plot(range(epochs), yList)
plt.xticks(range(epochs));
```

```python
plt.plot(range(epochs), errorList)
plt.xticks(range(epochs));
plt.xlabel("Epochs")
plt.ylabel("Error")
```
